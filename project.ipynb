```python
#Import all the libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#Taking dataset
foot_data = pd.read_csv('D:\\Study\\Data Science\\Football Project\\dataset.csv')

#Dataset
foot_data
```

```python
#use info() function for knowing the dataset null values, it will return null values as well as data type used for it.
foot_data.info()
```

```python
#use describe() function for knowing the row infromation, i.e. it return count, mean, std, min, max, etc.
foot_data.describe()
```

```python
#isna().sum() used for return null values from describe function
foot_data.isna().sum()
```

```python
#Create a new column as Minutes per match
foot_data['min_per_match'] = (foot_data['time'] / foot_data['games']).astype(int)

#Create a new column as Goals per match
foot_data['goals_per_match'] = (foot_data['goals'] / foot_data['games']).astype(float)

print(foot_data.head())
```

```python
#Median age of players
age = foot_data['age']

avg_age = np.median(age)
print(avg_age)
```

```python
#Total Goals, use goals column and use sum() function for addition of total goals scored by following players
total_goals = foot_data['goals'].sum()
print(total_goals)
```

```python
#Total penalty goals
penalty_goals = foot_data['penalty_goals'].sum()
print(penalty_goals)
```

```python
#Total penalty attempted
attempted_penalty = foot_data['penalty_attempts'].sum()
print(attempted_penalty)
```

```python
#Calculate average conversion rate
conversion_rate1 = foot_data['conversion_rate'].mean()
print(conversion_rate1)
```

```python
#Find unique numbers of positions
foot_data['position'].unique()
```

```python
#Total forward players, F=Forword & S=Side Forward(Wingers)
foot_data[foot_data['position'] == 'F, S']
```

```python
#Players from different countries
np.size((foot_data['Nationality'].unique()))
```

```python
#Most players comes form which nation
nationality = foot_data.groupby('Nationality').size().sort_values(ascending=False)
nationality.head(10).plot(kind='bar', figsize=(12, 6), color = sns.color_palette('magma'))
```

```python
#Clubs with most players in squad
foot_data['team_title'].value_counts().nlargest(5).plot(kind='bar', color = sns.color_palette('viridis'))
```

```python
#Total assists from each club

assists_by_club = pd.DataFrame(foot_data.groupby('team_title', as_index=False)['assists'].sum())
sns.set_theme(style='whitegrid', color_codes=True)
ax = sns.barplot(x='team_title', y = 'assists', data=assists_by_club.sort_values(by='assists'),palette='Set2')
plt.xticks(rotation=75)
plt.xlabel("Club", fontsize=30)
plt.ylabel("Assists", fontsize=30)
plt.title("Total Assists From Each Club")
plt.rcParams['figure.figsize'] = (20, 9)
```

```python
#Scatter diagram of expected Goals from each player.

goals = foot_data['goals'].head(10)
expected_goals = foot_data['xG'].head(10)
plt.xlabel("Goals")
plt.ylabel("Exceted Goals")
plt.title("Goals vs expected goals")

  
# plot data in grouped manner of bar type
plt.plot(goals, expected_goals)
plt.show()
```

```python
#clubs with maximum players in thier squad

foot_data['Nationality'].value_counts().nlargest(5).plot(kind='bar', color=sns.color_palette('viridis'))
```

```python
#clubs with minimum players in thier squad

foot_data['Nationality'].value_counts().nsmallest(5).plot(kind='bar', color=sns.color_palette('viridis'))
```

```python
#Players based on age group

under20 = foot_data[foot_data['age'] <= 20]
age20_25 = foot_data[(foot_data['age'] > 20) & (foot_data['age'] <= 25)]
age25_30 = foot_data[(foot_data['age'] > 25) & (foot_data['age'] <= 30)]
above30 = foot_data[foot_data['age'] > 30]
```

```python
x = np.array([under20['player_name'].count(), age20_25['player_name'].count(), age25_30['player_name'].count(), above30['player_name'].count()])
my_labels = ["<=20", ">20 & <=25", ">25 & <=30", ">30"]
plt.title("Total players in each age group")
plt.pie(x, labels = my_labels, autopct="%.1f%%")
plt.show()
```

```python
#Players age above 30

above30['team_title'].value_counts().plot(kind='bar', color = sns.color_palette('cubehelix'))
```

```python
#most goals by player

top_10_goals = foot_data[["player_name", 'team_title', 'games', 'goals']].nlargest(n=10, columns='goals')
top_10_goals
```

```python
#Top 10 Goal contributors

top_10_goal_contributors = foot_data[["player_name", 'team_title', 'games', 'goals', 'assists', 'total_contribution']].nlargest(n=10, columns='total_contribution')
top_10_goal_contributors
```
